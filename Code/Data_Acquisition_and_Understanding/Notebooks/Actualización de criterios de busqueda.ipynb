```python
import pandas as pd
```

```python
# Cargue de los insumos para el filtrado de los datos
criterios = pd.read_excel('..\..\..\Sample_Data\Raw\Criterios de busqueda\Criterios.xlsx')
criterios.head()
```

```python
# Se eliminan los espacios en blanco al inicio de los registros de la columna 'Nombre_libro'
criterios['Nombre_libro'] = criterios['Nombre_libro'].str.lstrip()
criterios.head()
```

```python
# Diccionario cuyas llaves son parejas de elementos (criterio, nombre_libro) y cuyos valores son los criterios de busqueda
criterios_dict = {}
for criterio in criterios['Criterio'].unique():
    for nombre_libro in criterios[criterios['Criterio'] == criterio]['Nombre_libro'].unique():
        criterios_dict[criterio, nombre_libro] = criterios[criterios['Criterio'] == criterio][criterios['Nombre_libro'] == nombre_libro]
```

```python
# Exploración de los criterios de busqueda y su cantidad de registros
for key, value in criterios_dict.items():
    print(key, len(value))
```

```python
# Se guardan los criterios de busqueda en archivos csv
for key, value in criterios_dict.items():
    value['Contenido'].to_csv(f'..\..\..\Sample_Data\Raw\Criterios de busqueda\{key[0]}\{key[1]}.csv', index=False)
```
